```python
import pandas as pd 
import numpy as np
df = pd.read_csv("JEOPARDY_CSV.csv")
df
```

How many rows are there in the dataframe? What does a row represent in this

There are 216930 rows and a row represents a jeopardy question , answer , show number,category, 
value and other relevant information.

Clean the data. How do you need to “clean” it? What is “dirty” about this data?

If we look at the data types it's quiet evident that date column and value column are also showing object data types so its necessary to convert them into proper data types .
For Air Date the converted datatype would datetime64[ns]  and for value it would be float. And also,some column names have spaces , special characters so i have replaced them as well.

```python
#For removing column spaces 
df.columns = df.columns.str.replace(' ', '')
print(df.columns)
```

```python
import datetime
import pandas as pd
import numpy as np
import random
df["AirDate"] = pd.to_datetime(df["AirDate"])
df["AirDate"].head()
```

```python
df.dtypes
```

How many unique shows were there each year?


```python
df['Year'] = pd.DatetimeIndex(df['AirDate']).year
df.head()
group_names = df.groupby("Year")
group_names["ShowNumber"].nunique()
```

How many times was each category used?


```python
df["Category"].value_counts()
```

Which month (over the years) has had the most shows?

```python
#create a column for the month aired
df["Month"] = pd.DatetimeIndex(df['AirDate']).month
df
```

```python
monthly_data = df.groupby(["Year", "Month"])
ShowNumber = monthly_data["ShowNumber"].nunique()

Dict = {
    "ShowNumber" :ShowNumber
}

month_over_years = pd.DataFrame(Dict)

month_over_years.set_index = None
month_over_years
#group_names[["Month" , "ShowNumber"]].nunique()
```
